```python
import numpy as np
```

```python
from sklearn import datasets
iris_data = datasets.load_iris()
```

```python
x = iris_data['data']
y = iris_data['target']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
```

```python
model = DecisionTreeClassifier()

# Define the k-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate the model using 10-fold cross validation
scores = cross_val_score(model, x, y, cv=kf)

# Print the average accuracy and standard deviation across all folds
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
model = DecisionTreeClassifier(max_depth=2)

# Define the k-fold cross validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Train and evaluate the model using 10-fold cross validation
scores = cross_val_score(model, x, y, cv=kf)

# Print the average accuracy and standard deviation across all folds
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.model_selection import LeaveOneOut
```

```python
model = DecisionTreeClassifier(max_depth=2)

# Define the k-fold cross validation
LOO = LeaveOneOut()

# Train and evaluate the model using 10-fold cross validation
scores = cross_val_score(model, x, y, cv=LOO)

# Print the average accuracy and standard deviation across all folds
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

# HW6 วนลูปทำ 10-fold cross validation

```python
dt1 = shuffled_df[shuffled_df['fold']==1]
dt2 = shuffled_df[shuffled_df['fold']==2]
dt3 = shuffled_df[shuffled_df['fold']==3]
dt4 = shuffled_df[shuffled_df['fold']==4]
dt5 = shuffled_df[shuffled_df['fold']==5]
dt6 = shuffled_df[shuffled_df['fold']==6]
dt7 = shuffled_df[shuffled_df['fold']==7]
dt8 = shuffled_df[shuffled_df['fold']==8]
dt9 = shuffled_df[shuffled_df['fold']==9]
dt10 = shuffled_df[shuffled_df['fold']==10]
```

```python
# Loop over the dataframes
data = [dt1,dt2,dt3,dt4,dt5,dt6,dt7,dt8,dt9,dt10]

for dt in data:

    # Split the dataframe into features and target
    a = shuffled_df.drop('label', axis=1)
    b = shuffled_df['label']

    # Create a decision tree classifier object
    dtc = DecisionTreeClassifier()

    # Define a 10-fold cross-validation object
    LOO = LeaveOneOut()

    # Compute the cross-validation scores
    scores = cross_val_score(dtc, a, b, cv=LOO)

    # Print the mean and standard deviation of the cross-validation scores
    print("Mean accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
